## Qwen 3B

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
model_name = "Qwen/Qwen2.5-3B-Instruct"

# Загрузка токенизатора
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Конфигурация 4-битной квантизации
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype="auto"
)
```

```python
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128
)

prompt = "This is a test prompt. Can you hear me?"
output = pipe(prompt)[0]['generated_text']
print(output)
```

```python
from gigachat import GigaChat
from google.colab import userdata
```

```python
# Укажите ваш токен (ни в коем случае не публикуйте его!)
token = userdata.get('GigaToken')

# Создаем клиент
client = GigaChat(credentials=token, verify_ssl_certs=False)

# Запрос на русском языке
response = client.chat("Столица Франции —")
print(response.choices[0].message.content)
```
